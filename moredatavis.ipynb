```python
import pandas as pd 
import math
import matplotlib.pyplot as plt
import numpy as np
import datetime

df = pd.read_csv('JS5min/plot_data.csv')
df2 = pd.read_csv('JS10min/plot_data.csv')
df3 = pd.read_csv('JS30min/plot_data.csv')
df4 = pd.read_csv('String5min/plot_data.csv')
df5 = pd.read_csv('String10min/plot_data.csv')
df6 = pd.read_csv('String30min/plot_data.csv')
```

```python
plt.plot(df['# unix_time']-df['# unix_time'][0],df[' unique_crashes'], label='5 min')
plt.plot(df2['# unix_time']-df2['# unix_time'][0],df2[' unique_crashes'], label='10 min')
plt.plot(df3['# unix_time']-df3['# unix_time'][0],df3[' unique_crashes'], label='30 min')
plt.legend()
plt.title('Unique crashes over time in JavaScript Generator')
plt.xlabel('Time')
plt.ylabel('# of Unique Crashes')
```

```python
x = [5, 10, 30]
y1 = [df[' unique_crashes'].values[len(df[' unique_crashes'])-1], df2[' unique_crashes'].values[len(df2[' unique_crashes'])-1], df3[' unique_crashes'].values[len(df3[' unique_crashes'])-1]]
y2 = [df4[' unique_crashes'].values[len(df4[' unique_crashes'])-1], df5[' unique_crashes'].values[len(df5[' unique_crashes'])-1], df6[' unique_crashes'].values[len(df6[' unique_crashes'])-1]]

width = 0.40
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, y1, width)
plt.bar(x+0.2, y2, width)
```
