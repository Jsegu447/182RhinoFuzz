```python
import pandas as pd 
import math
import matplotlib.pyplot as plt
import numpy as np
import datetime

df = pd.read_csv('String30min/plot_data.csv')
df2 = pd.read_csv('JS30min/plot_data.csv')
```

```python
plt.plot(df['# unix_time']-df['# unix_time'][0],df[' paths_total'], label='String Input')
plt.plot(df2['# unix_time']-df2['# unix_time'][0],df2[' paths_total'], label='JavaScript Generator Input')
plt.legend()
plt.title('Total paths covered over time (30 min)')
plt.xlabel('Time')
plt.ylabel('Total Paths Covered')
```

```python
plt.plot(df['# unix_time']-df['# unix_time'][0],df[' unique_crashes'], label='String Input')
plt.plot(df2['# unix_time']-df2['# unix_time'][0],df2[' unique_crashes'], label='JavaScript Generator Input')
plt.legend()
plt.title('Unique crashes over time (30 min)')
plt.xlabel('Time')
plt.ylabel('# of Unique Crashes')
```

```python
plt.plot(df['# unix_time']-df['# unix_time'][0],df[' valid_cov'].str[:-1].astype('double'), label='String Input')
plt.plot(df2['# unix_time']-df2['# unix_time'][0],df2[' valid_cov'].str[:-1].astype('double'), label='JavaScript Generator Input')
plt.legend()
plt.title('Valid coverage over time (30 min) ')
plt.xlabel('Time')
plt.ylabel('Percentage of valid coverage')
```

```python
plt.plot(df2['# unix_time']-df2['# unix_time'][0],df2[' valid_inputs'], label='Valid Inputs')
plt.plot(df2['# unix_time']-df2['# unix_time'][0],df2[' invalid_inputs'], label='Invalid Inputs')

plt.legend()
plt.title('Total inputs over time in JavaScript Generator (30 min)')
plt.xlabel('Time')
plt.ylabel('Number of inputs')
```
